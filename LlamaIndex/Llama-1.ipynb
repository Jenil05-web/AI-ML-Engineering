```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
documents = SimpleDirectoryReader(input_files=['1706.03762v7.pdf']).load_data()
```

```python
print(documents)
```

```python
index= VectorStoreIndex.from_documents(documents) # llama index used to create vector store
index
# in this cell basically we are creating a vector store index from the documents that we have loaded. The VectorStoreIndex is a data structure that allows us to efficiently store and query the documents based on their vector representations.
```

```python
query_engine = index.as_query_engine() # query engine is pre made function used to query index (readymade function)
response = query_engine.query("What is the transformers?")

from llama_index.core.response.pprint_utils import pprint_response
from llama_index.core.retrievers import VectorIndexRetriever
from llama_index.core.query_engine import RetrieverQueryEngine
from llama_index.core.indices.postprocessor import SimilarityPostprocessor
retriever = VectorIndexRetriever(index=index , similarity_top_k=3)
query_engine = RetrieverQueryEngine(retriever=retriever, node_postprocessors=[SimilarityPostprocessor(similarity_top_k=3)])
postprocessor = SimilarityPostprocessor(similarity_cutoff=0.80)
response = query_engine.query("What is the transformers?")
pprint_response(response, show_source=True)
```

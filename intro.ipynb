## Pydantic Basics : Creating and Using Models

Pydantic models are foundation of data validation in Python. They use Python type annotations to define the structure and validate data at runtime. Basically they set an set of rules or validations so that it could be formatted in that specific format only


```python
from pydantic import BaseModel
```

```python
class Person(BaseModel):
    name : str
    age : int
    city : str

person = Person(name="Jenil", age=15,city="Anand")

print(person)

```

```python
person2 = Person(name='jenil',age=12,city=12)
print(person2) ##  In this case you will get an error becacuse you have defined above Person class structure
```

## Model with Optional Fields

Add optional fields using Python's Optional type :

```python
from typing import Optional
class Employee(BaseModel):
    id : int
    name : str
    department : str
    salary: Optional[float]= None # Optional with default value
    is_active: Optional[bool]= True # optional with default True
```

## HuggingFace X Langchain

```python
from langchain_huggingface import HuggingFaceEndpoint
from langchain_huggingface import ChatHuggingFace
from langchain_core.messages import HumanMessage, SystemMessage

## Basically HuggingfaceEndpoint provides an interface to work with huggingface models through API without actaully installing them in system
```

```python
# Loading enviroment variables
import os
from dotenv import load_dotenv
load_dotenv()
hf_token = os.getenv("HUGGINGFACEHUB_API_TOKEN")
lc_key = os.getenv("LANGCHAIN_API_KEY")
```

```python
repo_id = "Qwen/Qwen2.5-7B-Instruct" # write the name of the model which we are using 

# create an endpoint to hugging face and define the llm model
llm = HuggingFaceEndpoint(
    repo_id=repo_id,
    task="text-generation", # always define the task 
    max_new_tokens=512,
    temperature=0.5,
    huggingfacehub_api_token=os.getenv("HUGGINGFACEHUB_API_TOKEN")
)
```

```python
chat_model = ChatHuggingFace(llm=llm)
messages = [
    HumanMessage(content="What is Langchain ?")
]
response = chat_model.invoke(messages)
print(response.content)
```

```python
from langchain_core.prompts import ChatPromptTemplate

chat_model = ChatHuggingFace(llm=llm)

prompt=ChatPromptTemplate.from_messages([("system", "You are a helpful assistant."),
    ("user", "Question:{question}"),])
final_chain = prompt | chat_model

response = final_chain.invoke({"question":"Who won cricket world cup of 2019 ?"})
print(response.content)

## The Golden Rule of LangChain Prompts & Models
## If you use PromptTemplate ➡️ Pipe it or chain to HuggingFaceEndpoint (llm)
## If you use ChatPromptTemplate ➡️ Pipe it or chain  to ChatHuggingFace (chat_model)
```
